 2. Educational level, with the categories: Elementary school, high school, college graduate, PHD ranked from 1 to 4

 When the categorical variables are ordinal , the most strightforward best approach is to replace the labels by some ordinal number based on the rank.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
# create a variable with dates, and from that extract the weekday
# create a list of dates with 20 days difference from today
# and then transform it into dataframe

df_base=datetime.datetime.today()
df_date_list=[df_base - datetime.timedelta(days=x) for x in range(0,20)]
df=pd.DataFrame(df_date_list)
df.columns=['day']
df
```

```python
# extract the week day name


df['day_ordinal']=df['day'].dt.weekday
df.head(20)  # 0 is Monday
```
