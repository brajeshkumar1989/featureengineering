### *Mutual Information*

Ml Estimate mutual information for a discrete variable.
Mutual Information (MI) between two random variables is a non-negative value, which measures the dependency between the variables. It is equal to zero if and only if two random variables are independent , and higher values means higher dependency.

The function relies on nonparametric methods based on entrophy estimation from K-nearest neighbours distances.

A quantity called mutual information measures the amount of information one can obtain random variable given another.

The mutual information between two random variables X and Y can be stated formaly as follows:

#### *I(X;Y) = H(X) - H(X|Y) where I(X;Y) is the mutual information for X and Y, H(X) is the entropy for X and H(X|Y) is the conditional entropy for X given Y. The result has the units of bits.*

```python
import pandas as pd

# dataset from https://gist.github.com/tijptjik/9408623

df=pd.read_csv("wine.csv")
df.head()
```

```python
df['Wine'].unique()
```

```python
df.info()
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['Wine'], axis=1), df['Wine'],test_size=0.3, random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.feature_selection import mutual_info_classif # it will not return -ve value

# determine the mutual information
mutual_info=mutual_info_classif(X_train,y_train)

# highest value is showing most important feature e.g. 0.71705972
mutual_info
```

```python
mutual_info= pd.Series(mutual_info)
mutual_info.index= X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
# plot the ordered mutual_info per feature
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20,8))
```

```python
from sklearn.feature_selection import SelectKBest # to select top 10 or some other number of important variables or features

# in thi example top 5 values got selected
sel_five_cols=SelectKBest(mutual_info_classif,k=5)
sel_five_cols.fit(X_train.fillna(0), y_train)
X_train.columns[sel_five_cols.get_support()]
```

#### **Difference between information Gain and Mutual information

I(X:Y)=H(X)-H(X|Y) and IG(S,a)= H(S)-H(S|a)

As such , mutual information is sometimes used as a synonym for information gain, technically they calculate the same quantity if applied to the same data.
