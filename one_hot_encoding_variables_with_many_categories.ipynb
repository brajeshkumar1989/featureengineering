```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
# check how many labels in each variable has

for col in data.columns:
  print(col, ':', len(data[col].unique()),'labels')
```

```python
# shape of the matrix without one hot encoding
data.shape
```

```python
# examine how many columns we will obtain after one hot encoding these variables
pd.get_dummies(data, drop_first=True).shape
```

```python
#find the top 10 most frequent categories for the variable x2

data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# make a list with the most frequent categories of the variable

top_10=[X for X in data.X2.value_counts().sort_values(ascending=False).head(10).index]

top_10
```

```python
# now make the 10 binary variables

for label in top_10:
  data[label] =np.where(data['X2']==label, 1,0)

data[['X2']+top_10].head(40)
```

```python
#get whole set of dummy variables, for all the categorical variables

def one_hot_top_x(df,variable,top_x_labels):
  #function to create the dummy variables for the most frequent labels
  #we can vary the number of most frequent labels that we encode

  for label in top_x_labels:
    df[variable+'_'+label]=np.where(data[variable]==label,1,0)

# read the data again
data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])

# encode X2 into the 10 most frequent categories
one_hot_top_x(data, 'X2',top_10)
data.head()
```

```python
# find the 10 most frequent categories for X1
top_10=[X for X in data.X1.value_counts().sort_values(ascending=False).head(10).index]

# now create the 10 most frequent dummy variables for X1
one_hot_top_x(data,'X1',top_10)
data.head()
```

### One Hot encoding of top variables

### Advantages

  1. Straightforward to implement
  2. Does not require hrs of variable exploration
  3. Does not expand massively the feature spacce (number of columns in the dataset)

### Disadvantages

 1. Does not add any information that may make the variable more predective
 2. Does not keep the information of the ignored labels


It is not unusual that categorical variables have few dominating categories and the remaining labels add mostly noise , this is a quite simple and strightforward approach that may be useful on many occasions.

top 10 is arbitrary number in this case.
