## **Probability Ratio Encoding**

#### Steps
1. Probability of Survived based on Cabin--- Categorical feature
2. Probability of not survived --- 1-(probability of survived)
3. pr(Survived)/pr(Not Survived)
4. Dictionary to map cabin with probability
5. Replace Probability ratio encoded value from point 4 to categorical feature Cabin

```python
import pandas as pd
```

```python
df=pd.read_csv('Titanic-Dataset.csv', usecols=['Cabin','Survived'])

df.head()
```

```python
### Replacing
df['Cabin'].fillna('Missing',inplace=True)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]

df.head()
```

```python
df.Cabin.unique()
```

```python
### group by Cabin and get mean for Survived column and store to variable
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']

prob_df.head()
```

```python
probability_encoded=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(probability_encoded)

df.head(20)
```
