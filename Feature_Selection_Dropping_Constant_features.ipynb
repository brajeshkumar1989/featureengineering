##### Remove variables having constant features those are not important to solve the given problem.

```python
import pandas as pd

#create dataframe for the given data

data=pd.DataFrame({
    "A":[1,2,4,1,2,4],
    "B":[4,5,6,7,8,9],
    "C":[0,0,0,0,0,0],
    "D":[1,1,1,1,1,1],
})
```

```python
data.head()
```

#### **Variance Threshold**

Feature selector that removes all low-variance features.

This feature selection algorithm looks only at the features(X) , not the desired outputs(y), and can thus can be used for unsupervised learning.

```python
#it will remove zero variance feature
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(data)

# true means variance present and false is variance not present.
var_thres.get_support()
```

```python
constant_columns=[column for column in data.columns if column not in data.columns[var_thres.get_support()]]

print(constant_columns)

for feature in constant_columns:
  print(feature)

#dropping constant columns from dataset
data.drop(constant_columns, axis=1)
```
