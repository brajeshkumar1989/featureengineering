## Encoding Theory

required for categorical variable
e.g.: Gender -> male, female

2 types of categorical variables

  1. Nominal encoding - no order matter e.g: male, female
  2. Ordinal encoding - order matters , based on Rank e.g Salary for different qualified people , rearrange the category if required.


### Nominal Encoding

techniques of encoding for Nominal type

 1. one hot encoding
 2. one hot encoding with many categorical
 3. Mean encoding

### Ordinal encoding

 techniques of encoding
  
  1. Label encoding
  2. Target guided ordinal encoding
  
### One Hot Encoding

e.g:
      state
     -------
     Germany
     France
     Spain


    Germany  France  Spain
    ----------------------
    1         0       0
    0         1       0
    0         0       1

can delete Spain column and can identify using rest of columns

for 5 categories there will be 5 columns

### Disadvantage of One Hot encoding

e.g pincode if i try to convert it to one hot encoding and if there are 100 unique values means it will create 100 columns  

### Label encoding (Ordinal)

e.g:

    education
    ---------
    BE        1
    MAS       3
    PHD       4
    STATS     2   

In above table 1,2,3,4 are labels assigned to each value , assign highest label value to highest important categorical value.

### One hot encoding with multiple categories

e.g. feature1 has 50 different categories , take top 10 (arbitrary ) categories got repeated more number of times (Nominal categories).


### Target guided ordinal category (Ordinal encoding)

not consider just independent variable but also dependent variable
e.g:

    f1    o/p   Mean
    -----------------
    A     1      0.75  -->4
    B     1      0.6   -->3
    C     0      0.4   -->2
    D     1      0.2   -->1
    A     0            -->4
    B     0            -->3

Based on each column find mean of o/p for specific category value like A values to calculate mean and assign rank based on calculated mean to each Category , for, e.g: here A has max mean value so assign highest value, applicable only for Ordinal.

### Mean encoding

    f1   o/p    mean

Best fit for Pincode , so in place of pincode replace them with mean values , unlike Ordinal where based on mean value weight we can assign a discreet value like 1,2,3,4

## Handle Missing values in categorical variables

  1. Delete the rows <br/>
    --if Dataset is less records then do not delete, this can be used for a large data set to delete 1 or 2 rows
   <br/>
  2. Replace with the most frequent values <br/>
    -- it could result imbalance dataset
    
  
  3. Apply classifier algoritm to predict (efficient way)<br/>
     
     e.g

          f1      f2  f3  f4  o/p
          -------------------------
          Male    23  12  22  Yes
          Female  20  10  12  Yes
          ---     10  10  12  NO
          Female  19  14  12  Yes

  In this example to identify the missing value
    a. consider all missing rows a test set and others as training set

    b. consider f2 till o/p columns as independent variables and f1 as dependent variable

    c. apply algoritm like KNN , or some classification algoritm to train a model based on train set and then pass test data to predict the values

    d. update the dataset for missing values for f1 and then this can be used as a dataset to predict the o/p

  4. Unsupervised learning (clustering technique)
    skip f1 columns completely and also o/p

    then create 2 categories as f1 has only two category Male and Female, and then use KNN to identify f1 value

## Ordinal measurement advantages

https://colab.research.google.com/drive/19lWbHa-IS9rQXTItVkRuwxnMa-TthGPU#scrollTo=EugsG_yPGs8_

Ordinal measurement is normally used for surveys and questionnaires. Statistical analysis is applied to the responses once they are collected to place the people who took the survey into the various categories. The data is then compared to draw inferences and conclusions about the whole surveyed population with regard to the specific variables. The advantage of using ordinal measurement is ease of collation and categoriation. If you ask a survay question without providing the variables, the answers are likely to be so diverse they cannot be converted to statistics.

with respect to Machine Learning

  1. Keeps the semantical information of the variable(Human readable content)
  2. Straightforward

## Ordinal Measurement Disadvantages

the same characteristics of ordinal measurement that create its advantages also create certain disadvantages. The response are often so narrow to the question that they create or magnify bias that is not factored into the servay. For example ,on the question about satisfaction with the governor, people might be satisfied with his job performance but upset about recent sex scandal. They survey question might lead respondents to state their dissatisfaction about the scandal in spite of satisfaction with his job performance- but the statistical conclusion will not differenciate.
